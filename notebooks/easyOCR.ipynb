# Bouldering segmentation
## Detect "Boulder n" to segment the video into sections

```python
from src.bouldering.media.video.video import Video
from src.bouldering.processing.scene import SceneSplitterOCR
```

```python
# Configuration
VIDEO_PATH = "../data/video.mp4"
LANGS = ["en"]
RESIZE_FX = .5
RESIZE_FY = .5
CROPBOX = [0, 0.5, 0.5, 1]
BATCH_SIZE = 16
DECODER = "greedy"  # "beamsearch" gives better accuracy but slower
MIN_CONF = 0.0  # keep all OCR outputs
USE_GPU = None
```

```python
vid = Video.read(VIDEO_PATH)
```

```python
vid = Video.read(VIDEO_PATH)

splitter = SceneSplitterOCR(
    langs=["en"],
    crop_box=[0, 0.5, 0.5, 1],
    stride=3,           # sample every 3 frames
    batch_size=16,
    fx=0.5, 
    fy=0.5,
    smooth_window=5,    # lookahead 5 sampled frames
    majority_ratio=0.60,
    require_number=True # only confirm when a number is parsed
)

starts = splitter.split(vid)  # or pass (x,y,w,h)
```

```python
Segments =[
    {'label': 'BOULDER 1', 'start_sec': 20.4, 'end_sec': 69.36},
    {'label': 'BOULDER 2', 'start_sec': 69.36, 'end_sec': 183.12},
    {'label': 'BOULDER 3', 'start_sec': 183.12, 'end_sec': 321.12},
    {'label': 'BOULDER 4', 'start_sec': 321.12, 'end_sec': 420.0}
]
```

```python
for i, segment in enumerate(segments):
    print(f"cut {i}")
    start = segment["start_sec"]
    end = segment["end_sec"]
    video_cut = vid.cut(start, end)
    video_cut.write(f"../data/boulder_{i}.mp4")
```
